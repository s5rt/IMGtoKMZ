```python
import zipfile
import xml.etree.ElementTree as ET

kmz_file = "bagapatia.kmz"
output_txt = "output.txt"

def dd_to_dms(dd, coord_type):
    dd = float(dd)
    direction = ''

    if coord_type == 'lat':
        direction = 'N' if dd >= 0 else 'S'
    elif coord_type == 'lon':
        direction = 'E' if dd >= 0 else 'W'

    dd = abs(dd)
    degrees = int(dd)
    minutes_float = (dd - degrees) * 60
    minutes = int(minutes_float)
    seconds = (minutes_float - minutes) * 60

    return f"{degrees}Â°{minutes}'{seconds:.2f}\"{direction}"

# Open KMZ
with zipfile.ZipFile(kmz_file, 'r') as kmz:
    kml_filename = [f for f in kmz.namelist() if f.endswith('.kml')][0]
    kml_data = kmz.read(kml_filename)

# Parse KML
root = ET.fromstring(kml_data)
ns = {'kml': 'http://www.opengis.net/kml/2.2'}
placemarks = root.findall('.//kml:Placemark', ns)

with open(output_txt, 'w', encoding='utf-8') as f:
    for placemark in placemarks:
        name = placemark.find('kml:name', ns)
        coords = placemark.find('.//kml:coordinates', ns)

        if name is not None and coords is not None:
            lon, lat, *_ = coords.text.strip().split(',')

            lat_dms = dd_to_dms(lat, 'lat')
            lon_dms = dd_to_dms(lon, 'lon')

            f.write(f"{name.text}, {lat_dms}, {lon_dms}\n")

print("Export completed successfully in DMS format.")
```
